```python
from telegram import Update
from telegram.ext import Application, CommandHandler, CallbackContext, ConversationHandler, MessageHandler, filters
import sqlite3
import logging
import nest_asyncio

# –ü—Ä–∏–º–µ–Ω—è–µ–º nest_asyncio –¥–ª—è Jupyter/Colab
nest_asyncio.apply()

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
conn = sqlite3.connect('naryads.db')
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS naryads
             (id INTEGER PRIMARY KEY,
              cadet_name TEXT,
              direction TEXT,
              count INTEGER)''')
conn.commit()

# –°–æ—Å—Ç–æ—è–Ω–∏—è –¥–ª—è ConversationHandler
DIRECTION, COUNT = range(2)

logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)

async def start(update: Update, context: CallbackContext):
    await update.message.reply_text(
        "–ü—Ä–∏–≤–µ—Ç! –Ø –±–æ—Ç –¥–ª—è —É—á—ë—Ç–∞ –Ω–∞—Ä—è–¥–æ–≤.\n"
        "–î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–æ–º–∞–Ω–¥—ã:\n"
        "/add - –î–æ–±–∞–≤–∏—Ç—å –Ω–∞—Ä—è–¥\n"
        "/stats - –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞\n"
        "/list - –°–ø–∏—Å–æ–∫ –≤—Å–µ—Ö –Ω–∞—Ä—è–¥–æ–≤"
    )

async def add_naryad_start(update: Update, context: CallbackContext):
    directions = ["–°—Ç—Ä–æ–µ–≤–∞—è", "–ö–∞–ø—Ä–∞–ª", "–•–æ–∑—Ä–∞–±–æ—Ç—ã", "–ö–ü–ü"]
    keyboard = [[(d, d)] for d in directions]
    await update.message.reply_text('–í—ã–±–µ—Ä–∏—Ç–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ:', reply_markup={'keyboard': keyboard, 'one_time_keyboard': True})
    return DIRECTION

async def add_naryad_direction(update: Update, context: CallbackContext):
    direction = update.message.text
    context.user_data['direction'] = direction
    await update.message.reply_text(f"–ù–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ: {direction}\n–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞—Ä—è–¥–æ–≤:")
    return COUNT

async def add_naryad_count(update: Update, context: CallbackContext):
    count = update.message.text
    if not count.isdigit():
        await update.message.reply_text("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ!")
        return COUNT

    direction = context.user_data['direction']
    cadet_id = update.message.from_user.id

    # –î–æ–±–∞–≤–ª—è–µ–º –≤ –±–∞–∑—É –¥–∞–Ω–Ω—ã—Ö
    cursor.execute("INSERT INTO naryads (cadet_name, direction, count) VALUES (?, ?, ?)",
                  (cadet_id, direction, int(count)))
    conn.commit()

    await update.message.reply_text(f"‚úÖ –î–æ–±–∞–≤–ª–µ–Ω–æ {count} –Ω–∞—Ä—è–¥–æ–≤ –ø–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é {direction}")
    return ConversationHandler.END

async def stats(update: Update, context: CallbackContext):
    cursor.execute("SELECT direction, SUM(count) FROM naryads GROUP BY direction")
    stats = cursor.fetchall()

    response = "üìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –Ω–∞—Ä—è–¥–æ–≤:\n"
    for row in stats:
        response += f"{row[0]}: {row[1]}\n"

    await update.message.reply_text(response)

async def cancel(update: Update, context: CallbackContext):
    await update.message.reply_text('–û–ø–µ—Ä–∞—Ü–∏—è –æ—Ç–º–µ–Ω–µ–Ω–∞')
    return ConversationHandler.END

def main():
    # –°–æ–∑–¥–∞—ë–º –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ –∏ –ø–µ—Ä–µ–¥–∞—ë–º —Ç–æ–∫–µ–Ω
    application = Application.builder().token("7309960131:AAG5l2S7QokOeH6BZkMerdY5tqXTWMcMhTA").build()

    # –û–±—Ä–∞–±–æ—Ç—á–∏–∫–∏ –∫–æ–º–∞–Ω–¥
    application.add_handler(CommandHandler("start", start))
    application.add_handler(CommandHandler("stats", stats))

    # ConversationHandler –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –Ω–∞—Ä—è–¥–æ–≤
    conv_handler = ConversationHandler(
        entry_points=[CommandHandler('add', add_naryad_start)],
        states={
            DIRECTION: [MessageHandler(filters.TEXT & ~filters.COMMAND, add_naryad_direction)],
            COUNT: [MessageHandler(filters.TEXT & ~filters.COMMAND, add_naryad_count)]
        },
        fallbacks=[CommandHandler('cancel', cancel)]
    )
    application.add_handler(conv_handler)

    # –ó–∞–ø—É—Å–∫ –±–æ—Ç–∞
    application.run_polling()

if __name__ == '__main__':
    main()
```
